Compute the mean , the median and the mode of chol

```python

print("Mean:", df["chol"].mean()) #mean is the average
print("Median:", df["chol"].median())  #sort , display the value in the middle , if the number of elements is pair , it calculates the average of both
print("Mode:", df["chol"].mode()[0])  #the most occured value
```

Calculate the range , Q1 and Q3 for age using numpy.percentile

```python

import numpy as np

age = df["age"]

# Calculate Q1 and Q3
Q1 = np.percentile(age, 25)
Q3 = np.percentile(age, 75)

# Calculate range
range = age.max() - age.min()

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("Range:", range)
```

Detect outliers

```python

Q1 = df['chol'].quantile(0.25)
Q3 = df['chol'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers
outliers = df[(df['chol'] < Q1 - 1.5 * IQR) | (df['chol'] > Q3 + 1.5 * IQR)]
print(outliers)
```
